```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as imageLoader
```

```python
data = np.random.randint(low = 0, high = 255, size = 24)

# image1
image = data.reshape(4, 6)

plt.figure(figsize=(12,8))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.xlabel(str(image.ndim) +'D '+ 'Gray scale image(4X6)')

# image2
image2 = data.reshape(2, 4, 3)
plt.subplot(1, 3, 2)
plt.imshow(image2)
plt.xlabel(str(image2.ndim) +'D - '+ ' Color image(2X4X3)')


# image3
image3 = data.reshape(2, 3, 4)
plt.subplot(1, 3, 3)
plt.imshow(image3)
plt.xlabel(str(image3.ndim) +'D - '+ ' Color image(2X3X4)')

```
