## DIAGONAL FINDING

```python
#diagonal
data = data.reshape(6, 6)
print(data)

print('Diagonal => ', data.diagonal())

#using fancy indexing
i = [0, 1, 2, 3, 4, 5]
print('Diagonal => ', data[i, i])

#insert new value in diagonal
data[i, i] = 0
print(data)

#upper diagonal
i2 = array([0, 1, 2, 3, 4])
print('upper diagonal =>', data[i2, i2+1])


#lower diagonal
print(data[i2+1, i2])
```

## ARRAY CALCULATOR METHOD
